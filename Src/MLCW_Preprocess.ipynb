```python
import pandas as pd
import os

from google.colab import drive
drive.mount('/content/drive')

# Step 1: Specify the file name in the same folder
file_name = "bank-additional-full.csv"  # Input file
output_file = "cleaned_file.csv"  # Output file

# Step 2: Check if the file exists in the current directory
if os.path.isfile(file_name):
    print(f"'{file_name}' found. Loading file...")

    # Step 3: Load the CSV file (assuming values are semicolon-separated)
    data = pd.read_csv(file_name, header=None)

    # Step 4: Check the first few rows to confirm the data format
    print("Raw data preview:")
    print(data.head())

    # Step 5: Split the single column into multiple columns based on semicolons
    data_split = data[0].str.split(';', expand=True)

    # Step 6: Remove double quotes from string values
    data_cleaned = data_split.applymap(lambda x: x.strip('"') if isinstance(x, str) else x)

    # Step 7: Display the cleaned data
    print("\nCleaned data preview:")
    print(data_cleaned.head())

    # Save the processed dataset to Google Drive
    data_cleaned.to_csv('/content/drive/My Drive/ML CW/data_cleaned.csv', index=False)

else:
    print(f"Error: '{file_name}' not found in the current directory.")
```
